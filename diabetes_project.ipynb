```python
# 1. Import libraries
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
# 2. Load dataset and convert to DataFrame
diabetes = load_diabetes()
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
```

```python
# 3. Split dataset into training and testing
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
# 4. Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("Linear Regression R²:", r2_score(y_test, y_pred_lr))
```

```python
# 5. Ridge Regression
ridge = Ridge(alpha=0.001)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
print("Ridge Regression R²:", r2_score(y_test, y_pred_ridge))
```

```python
# 6. RandomForest Regressor
rf = RandomForestRegressor(max_depth=10, n_estimators=500, random_state=55)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("RandomForest R²:", r2_score(y_test, y_pred_rf))
```
